```python
FSLE_url = "http://tds.aviso.altimetry.fr/thredds/dodsC/dataset-duacs-dt-global-allsat-madt-fsle"
```

```python
import requests
import xarray as xr

session = requests.Session()
session.auth = ('dhruvbalwada@gmail.com', 'fM0l0F')

store = xr.backends.PydapDataStore.open(FSLE_url,
                                        session=session)
```

```python
data_dir ="/Users/dhruvbalwada/OneDrive/sogos_data/"
data_folder = data_dir + "data/raw/Background Data/"
SSH_fname = data_folder + "SSH region 20190429 20190725.nc"
ds_SSH = xr.open_dataset(SSH_fname)
```

```python
ds = xr.open_dataset(store)
```

```python
ds
```

```python
FSLE_map= ds.sel(lon=slice(ds_SSH.longitude.min(),ds_SSH.longitude.max())
                       , lat=slice(ds_SSH.latitude.min(),ds_SSH.latitude.max())
                      ).sel(time=slice(ds_SSH.time.min(), ds_SSH.time.max()))
FSLE_map.load();
```
